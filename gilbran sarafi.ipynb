```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = {
    "Nama": ["Ali", "Budi", "Citra", "Dina", "Eko", "Fani", "Gilang", "Hana", "Irfan", "Joko"],
    "Umur": [10, 11, 10, 12, 11, 10, 12, 11, 10, 12],
    "Kelas": [4, 5, 4, 6, 5, 4, 6, 5, 4, 6],
    "Nilai_Matematika": [85, 78, 90, 65, 75, 88, 60, 82, 91, 70],
    "Nilai_Bahasa": [88, 80, 92, 70, 76, 85, 65, 79, 95, 72]
}
```

```python
df = pd.DataFrame(data)
```

```python
print("Data Murid SD:\n")
```

```python
print(df)
```

```python
print("\nStatistik Deskriptif:")
```

```python
print(df.describe())
```

```python
print("\nRata-rata nilai Matematika:", df["Nilai_Matematika"].mean())
```

```python
print("Rata-rata nilai Bahasa:", df["Nilai_Bahasa"].mean())
```

```python
print("\nJumlah Murid per Kelas:")
```

```python
print(df["Kelas"].value_counts())
```

```python
print("\nMurid dengan Nilai Bahasa Tertinggi:")
```

```python
print(df[df["Nilai_Bahasa"] == df["Nilai_Bahasa"].max()])
```

```python
plt.figure(figsize=(6,4))
plt.bar(df["Nama"], df["Nilai_Matematika"], color="skyblue")
plt.title("Nilai Matematika per Murid")
plt.xlabel("Nama")
plt.ylabel("Nilai")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
mean_per_kelas = df.groupby("Kelas")[["Nilai_Matematika", "Nilai_Bahasa"]].mean()
mean_per_kelas.plot(kind="bar", figsize=(7,5))
plt.title("Rata-rata Nilai per Kelas")
plt.ylabel("Nilai Rata-rata")
plt.xlabel("Kelas")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data = {
    "Nama": ["Ali", "Budi", "Citra", "Dina", "Eko", "Fani", "Gilang", "Hana", "Irfan", "Joko"],
    "Umur": [10, 11, 10, 12, 11, 10, 12, 11, 10, 12],
    "Kelas": [4, 5, 4, 6, 5, 4, 6, 5, 4, 6],
    "Nilai_Matematika": [85, 78, 90, 65, 75, 88, 60, 82, 91, 70],
    "Nilai_Bahasa": [88, 80, 92, 70, 76, 85, 65, 79, 95, 72]
}
df = pd.DataFrame(data)
```

```python
corr = df.drop(columns=["Nama"]).corr()
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(corr, annot=True, cmap="magma", fmt=".2f", linewidths=0.5)
plt.title("Heatmap Korelasi Antar Nilai dan Umur")
plt.show()
```
